## ğŸ—“ï¸ 4. Today's NBA Schedule

```python
import pandas as pd
from datetime import date

print("ğŸ€ Today's NBA Games\n")
print("="*60)

try:
    from qepc.sports.nba.sim import get_today_games
    
    today_games = get_today_games(show='clean')
    
    if today_games is None or today_games.empty:
        print(f"\nğŸ“­ No games scheduled for {date.today().strftime('%B %d, %Y')}")
    else:
        print(f"\nğŸ¯ {len(today_games)} game(s) today:\n")
        
        for idx, game in today_games.iterrows():
            away = game.get('Away Team', 'TBD')
            home = game.get('Home Team', 'TBD')
            time = game.get('Time', 'TBD')
            
            print(f"   ğŸ€ {away} @ {home}")
            print(f"      ğŸ• {time}\n")
            
except Exception as e:
    print(f"âš ï¸  Could not load schedule: {e}")
    print("   Check that data/Games.csv exists")

print("="*60)
```

## ğŸ¯ 5. Quick Prediction (Single Game)

```python
# Customize these teams!
AWAY_TEAM = "Boston Celtics"
HOME_TEAM = "Los Angeles Lakers"

print(f"ğŸ”® Quick Prediction\n")
print("="*60)
print(f"\nğŸ€ {AWAY_TEAM} @ {HOME_TEAM}\n")

try:
    from qepc.sports.nba.sim import simulate_game
    
    result = simulate_game(
        home_team=HOME_TEAM,
        away_team=AWAY_TEAM
    )
    
    print("ğŸ“Š Prediction:")
    print(f"   Expected Score: {result['home_score']:.1f} - {result['away_score']:.1f}")
    print(f"   Win Probability: {HOME_TEAM}: {result['home_win_prob']:.1%}")
    print(f"   Predicted Total: {result['total']:.1f} points")
    print(f"   Confidence: {'High' if abs(result['home_win_prob'] - 0.5) > 0.2 else 'Medium'}")
    
except Exception as e:
    print(f"âš ï¸  Prediction failed: {e}")
    print("   Make sure team names are spelled correctly")

print("\n" + "="*60)
```
