## ğŸ”¬ 3. Test QEPC Imports

```python
print("ğŸ§ª Testing QEPC module imports...\n")

import_errors = []

def test_import(label, import_func):
    """Test a single import and report status"""
    try:
        import_func()
        print(f"âœ… {label}")
        return True
    except Exception as e:
        print(f"âŒ {label}: {type(e).__name__}")
        import_errors.append((label, e))
        return False

# Test each major module
test_import("Autoload", lambda: __import__('qepc_autoload'))
test_import("Lambda Engine", lambda: __import__('qepc.core.lambda_engine', fromlist=['compute_lambda']))
test_import("Simulator", lambda: __import__('qepc.core.simulator', fromlist=['run_qepc_simulation']))
test_import("NBA Sim", lambda: __import__('qepc.sports.nba.sim', fromlist=['get_today_games']))
test_import("Backtest Engine", lambda: __import__('qepc.backtest.backtest_engine', fromlist=['run_season_backtest']))
test_import("Strengths V2", lambda: __import__('qepc.sports.nba.strengths_v2', fromlist=['calculate_advanced_strengths']))

# Summary
print("\n" + "="*60)
if import_errors:
    print(f"âš ï¸  {len(import_errors)} import(s) failed")
    print("\nFailed imports:")
    for label, error in import_errors:
        print(f"   â€¢ {label}")
else:
    print("âœ… All QEPC modules imported successfully!")
print("="*60)
```

## ğŸ—“ï¸ 4. Today's NBA Schedule

```python
import pandas as pd
from datetime import datetime, timedelta

print("ğŸ€ Today's NBA Games\n")
print("="*60)

# Load schedule
games_path = project_root / "data" / "Games.csv"
df = pd.read_csv(games_path)

# Convert dates
df['Date'] = pd.to_datetime(df['Date'])
today = datetime.today().date()

# Try today first
today_games = df[df['Date'].dt.date == today]

if today_games.empty:
    print(f"ğŸ“­ No games scheduled for {today.strftime('%B %d, %Y')}")
    print("   (NBA day off - likely Thanksgiving, Christmas, or All-Star break)\n")
    
    # Find next game day
    future_games = df[df['Date'].dt.date > today].sort_values('Date')
    
    if not future_games.empty:
        next_date = future_games['Date'].dt.date.iloc[0]
        next_games = df[df['Date'].dt.date == next_date]
        
        print(f"ğŸ”® Next games on {next_date.strftime('%A, %B %d')}:\n")
        
        for idx, game in next_games.iterrows():
            away = game['Away Team']
            home = game['Home Team']
            time = game['Time']
            print(f"   ğŸ€ {away} @ {home}")
            print(f"      ğŸ• {time}\n")
    else:
        print("âš ï¸  No upcoming games in schedule")
        
else:
    print(f"âœ… {len(today_games)} game(s) today:\n")
    
    for idx, game in today_games.iterrows():
        away = game['Away Team']
        home = game['Home Team']
        time = game['Time']
        print(f"   ğŸ€ {away} @ {home}")
        print(f"      ğŸ• {time}\n")

print("="*60)
```

## ğŸ¯ 5. Quick Single-Game Prediction

```python
"""
Fixed Quick Prediction Cell - Handles data/raw/ path
Replace your prediction cell with this
"""

# ğŸ¯ Quick Single-Game Prediction (FIXED PATH)

import pandas as pd
from pathlib import Path

# Customize these teams!
AWAY_TEAM = "Boston Celtics"
HOME_TEAM = "Los Angeles Lakers"

print(f"ğŸ”® Quick Prediction\n")
print("="*60)
print(f"\nğŸ€ {AWAY_TEAM} @ {HOME_TEAM}\n")

try:
    # Step 1: Load team stats manually (from correct location)
    print("ğŸ“Š Loading team strengths...")
    
    # Try data/raw/ first, then data/
    team_stats_paths = [
        project_root / "data" / "raw" / "Team_Stats.csv",
        project_root / "data" / "Team_Stats.csv",
        project_root / "data" / "raw" / "TeamStatistics.csv",
        project_root / "data" / "TeamStatistics.csv",
    ]
    
    team_stats = None
    for path in team_stats_paths:
        if path.exists():
            print(f"   Found at: {path}")
            team_stats = pd.read_csv(path)
            break
    
    if team_stats is None:
        raise FileNotFoundError(
            f"Could not find Team_Stats.csv in:\n" +
            "\n".join(f"   â€¢ {p}" for p in team_stats_paths)
        )
    
    print(f"   Loaded {len(team_stats)} teams")
    
    # Check if teams exist
    team_col = 'Team' if 'Team' in team_stats.columns else 'Tm'
    
    if AWAY_TEAM not in team_stats[team_col].values:
        print(f"\nâŒ Team not found: {AWAY_TEAM}")
        print(f"\nğŸ“‹ Available teams:")
        for team in sorted(team_stats[team_col].unique()):
            print(f"   â€¢ {team}")
        raise ValueError(f"Unknown team: {AWAY_TEAM}")
    
    if HOME_TEAM not in team_stats[team_col].values:
        print(f"\nâŒ Team not found: {HOME_TEAM}")
        print(f"\nğŸ“‹ Available teams:")
        for team in sorted(team_stats[team_col].unique()):
            print(f"   â€¢ {team}")
        raise ValueError(f"Unknown team: {HOME_TEAM}")
    
    # Get team data
    away_stats = team_stats[team_stats[team_col] == AWAY_TEAM].iloc[0]
    home_stats = team_stats[team_stats[team_col] == HOME_TEAM].iloc[0]
    
    # Step 2: Calculate strengths (simple version)
    print("âš™ï¸  Calculating matchup...")
    
    # Get offensive/defensive ratings (column names vary)
    def get_rating(stats, rating_type):
        """Get ORtg or DRtg from various column name formats"""
        possible_names = {
            'ORtg': ['ORtg', 'OffRtg', 'O_Rtg', 'Offensive Rating'],
            'DRtg': ['DRtg', 'DefRtg', 'D_Rtg', 'Defensive Rating']
        }
        
        for col_name in possible_names[rating_type]:
            if col_name in stats.index:
                return float(stats[col_name])
        
        # Fallback: estimate from pace and other stats if available
        if rating_type == 'ORtg':
            # Try to calculate from PTS and Pace
            if 'PTS' in stats.index and 'Pace' in stats.index:
                return float(stats['PTS']) / float(stats['Pace']) * 100
            return 110.0  # League average fallback
        else:  # DRtg
            return 110.0  # League average fallback
    
    away_ortg = get_rating(away_stats, 'ORtg')
    away_drtg = get_rating(away_stats, 'DRtg')
    home_ortg = get_rating(home_stats, 'ORtg')
    home_drtg = get_rating(home_stats, 'DRtg')
    
    print(f"\nğŸ“Š Team Ratings:\n")
    print(f"   {AWAY_TEAM}:")
    print(f"      Offensive Rating: {away_ortg:.1f}")
    print(f"      Defensive Rating: {away_drtg:.1f}")
    print(f"      Net Rating: {away_ortg - away_drtg:+.1f}")
    
    print(f"\n   {HOME_TEAM}:")
    print(f"      Offensive Rating: {home_ortg:.1f}")
    print(f"      Defensive Rating: {home_drtg:.1f}")
    print(f"      Net Rating: {home_ortg - home_drtg:+.1f}")
    
    # Step 3: Calculate expected performance
    print(f"\nğŸ² Running quick calculation...")
    
    # Expected points (simplified lambda approach)
    # Away team expected scoring rate vs home defense
    away_lambda = (away_ortg * home_drtg) / (110.0 ** 2) * 100
    
    # Home team expected scoring rate vs away defense (with home advantage)
    home_lambda = (home_ortg * away_drtg) / (110.0 ** 2) * 100 * 1.03
    
    # Estimate win probability using log5 formula
    away_strength = away_ortg - away_drtg
    home_strength = home_ortg - home_drtg + 3.0  # +3 for home court
    
    # Convert to probabilities
    home_win_prob = 1 / (1 + 10 ** ((away_strength - home_strength) / 25))
    away_win_prob = 1 - home_win_prob
    
    # Expected scores (rough estimate)
    expected_home = home_lambda
    expected_away = away_lambda
    expected_total = expected_home + expected_away
    expected_spread = expected_home - expected_away
    
    print(f"\nğŸ“Š Prediction Results:")
    print("-" * 60)
    
    print(f"\nğŸ¯ Win Probability:")
    print(f"   {HOME_TEAM}: {home_win_prob:.1%}")
    print(f"   {AWAY_TEAM}: {away_win_prob:.1%}")
    
    print(f"\nğŸ“ˆ Expected Score:")
    print(f"   {HOME_TEAM}: {expected_home:.1f}")
    print(f"   {AWAY_TEAM}: {expected_away:.1f}")
    print(f"   Total: {expected_total:.1f} points")
    print(f"   Spread: {HOME_TEAM} {expected_spread:+.1f}")
    
    # Confidence assessment
    confidence_diff = abs(home_win_prob - 0.5)
    if confidence_diff > 0.20:  # >70% or <30%
        confidence = "ğŸ”´ HIGH"
        desc = "Strong favorite"
    elif confidence_diff > 0.10:  # >60% or <40%
        confidence = "ğŸŸ¡ MEDIUM"
        desc = "Slight favorite"
    else:  # Between 40-60%
        confidence = "ğŸŸ¢ LOW"
        desc = "Toss-up game"
    
    print(f"\nğŸ’¡ Confidence Level: {confidence} ({desc})")
    
    # Winner prediction
    winner = HOME_TEAM if home_win_prob > 0.5 else AWAY_TEAM
    margin = abs(expected_spread)
    print(f"\nğŸ† Predicted Winner: {winner} by {margin:.1f} points")
    
    print(f"\nâš¡ Note: This is a quick estimate. For full quantum simulation")
    print(f"   with 100,000 states, use qepc_dashboard.ipynb or qepc_sandbox.ipynb")

except FileNotFoundError as e:
    print(f"âŒ File not found: {e}")
    print("\nğŸ’¡ Check that these files exist:")
    print("   â€¢ data/raw/Team_Stats.csv")
    print("   â€¢ data/Team_Stats.csv")
    
except KeyError as e:
    print(f"âŒ Column missing: {e}")
    print("\nğŸ’¡ Your Team_Stats.csv might have different column names")
    print("   Expected: Team, ORtg, DRtg")
    
except Exception as e:
    print(f"âŒ Prediction failed: {type(e).__name__}: {e}")
    print("\nğŸ’¡ Troubleshooting:")
    print("   1. Check team names match exactly (case-sensitive)")
    print("   2. Ensure Team_Stats.csv is in data/raw/ folder")
    print("   3. Check file has ORtg and DRtg columns")

print("\n" + "="*60)
```
