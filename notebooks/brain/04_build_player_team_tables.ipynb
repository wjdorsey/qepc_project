```python
import sys
from pathlib import Path

# Notebook lives in: <project_root>/notebooks/brain
PROJECT_ROOT = Path.cwd().resolve().parents[1]

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print("PROJECT_ROOT:", PROJECT_ROOT)
print("qepc in root?", (PROJECT_ROOT / "qepc").exists())
```

```python
from qepc.brain.teams_loader import (
    fetch_teams_static,
    fetch_league_team_season_stats,
)
from qepc.brain.players_loader import (
    fetch_players_static,
    fetch_league_player_season_stats,
)

# For now, use a completed season for stability. Once this works, we can switch to "2025-26".
season = "2023-24"

teams_df = fetch_teams_static()
players_df = fetch_players_static()

team_stats_df = fetch_league_team_season_stats(season, measure_type="Base")
player_stats_df = fetch_league_player_season_stats(season, measure_type="Base")

display(teams_df.head())
print("teams_df rows:", len(teams_df))

display(players_df.head())
print("players_df rows:", len(players_df))

display(team_stats_df.head())
print("team_stats_df rows:", len(team_stats_df))

display(player_stats_df.head())
print("player_stats_df rows:", len(player_stats_df))
```

```python
from pathlib import Path

out_dir = PROJECT_ROOT / "data" / "raw" / "nba"
(out_dir / "teams").mkdir(parents=True, exist_ok=True)
(out_dir / "players").mkdir(parents=True, exist_ok=True)

teams_df.to_parquet(out_dir / "teams" / "teams_static.parquet", index=False)
players_df.to_parquet(out_dir / "players" / "players_static.parquet", index=False)

team_stats_df.to_parquet(out_dir / "teams" / f"teams_stats_{season}_base.parquet", index=False)
player_stats_df.to_parquet(out_dir / "players" / f"players_stats_{season}_base.parquet", index=False)

print("Saved team & player tables to:", out_dir)
```
