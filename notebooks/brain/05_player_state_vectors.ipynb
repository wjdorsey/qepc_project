```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve().parents[1]

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print("PROJECT_ROOT:", PROJECT_ROOT)
print("qepc in root?", (PROJECT_ROOT / "qepc").exists())
```

```python
from qepc.brain.players_loader import fetch_league_player_season_stats
from qepc.brain.state_vectors import build_basic_player_state_vectors

season = "2023-24"  # we can switch to "2025-26" once we're happy

player_stats_df = fetch_league_player_season_stats(season, measure_type="Base")
print("player_stats_df shape:", player_stats_df.shape)

player_state_df = build_basic_player_state_vectors(player_stats_df)
print("player_state_df shape:", player_state_df.shape)

display(player_state_df.head())
```

```python
from pathlib import Path

out_dir = PROJECT_ROOT / "data" / "processed" / "nba" / "player_state_vectors"
out_dir.mkdir(parents=True, exist_ok=True)

player_state_df.to_parquet(out_dir / f"player_state_vectors_{season}_basic.parquet", index=False)

print("Saved player_state_vectors to:", out_dir)
```
