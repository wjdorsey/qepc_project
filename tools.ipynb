# **Lock File**
------

```python
import os
import stat

# The file you want to lock
target_file = "notebook_context.py" 

# Set permissions to Read-Only (444)
os.chmod(target_file, stat.S_IREAD)

print(f"ðŸ”’ {target_file} is now LOCKED (Read-Only).")
```

# **Unlock File**
___

```python
import os
import stat

# The file you want to unlock
target_file = "notebook_context.py"

# Set permissions to Read + Write (644)
os.chmod(target_file, stat.S_IREAD | stat.S_IWRITE)

print(f"ðŸ”“ {target_file} is now UNLOCKED (Writable).")
```
